```python
from google import genai
import re
import requests

def get_github_readme_content(github_url):
    """
    从 GitHub 项目链接中提取 README 的原始内容。
    支持常见的 GitHub URL 格式。会尝试 main 和 master 分支。
    """
    match = re.match(r"https://github\.com/([^/]+)/([^/]+)", github_url)
    if not match:
        print(f"错误：无法从 '{github_url}' 中解析 owner/repo。")
        return None

    owner, repo = match.groups()
    repo_cleaned = repo.replace(".git", "")
    
    readme_filenames = ["README.md", "README.rst", "README.txt", "readme.md"]
    default_branches = ["main", "master"] 
    
    content = None
    readme_url_used = None

    for branch in default_branches:
        for filename in readme_filenames:
            raw_url = f"https://raw.githubusercontent.com/{owner}/{repo_cleaned}/{branch}/{filename}"
            try:
                print(f"正在尝试获取 README 从: {raw_url}")
                response = requests.get(raw_url, timeout=10)
                response.raise_for_status() 
                content = response.text
                readme_url_used = raw_url
                print(f"成功获取 README 内容从: {readme_url_used}")
                return content 
            except requests.exceptions.HTTPError as e:
                if e.response.status_code == 404:
                    print(f"未找到 {filename} at {raw_url} (分支: {branch})。")
                else:
                    print(f"请求 README '{raw_url}' 时发生 HTTP 错误: {e.response.status_code} {e.response.reason}")
            except requests.exceptions.Timeout:
                print(f"请求 README '{raw_url}' 超时。")
            except requests.exceptions.RequestException as e:
                print(f"请求 README '{raw_url}' 时发生错误: {e}")
        if content: 
            break
            
    if not content:
        print(f"错误：在 {github_url} 中找不到任何常见的 README 文件 (已尝试分支: {', '.join(default_branches)})。")
        print("请检查链接或项目结构。")
    return content

client = genai.Client(api_key="AIzaSyB9pzbfTvYdFXcqoKEqi-Hx5QhHe6cCg6w")
github_project_url = "https://github.com/Mercerai/EvGGS"
readme = get_github_readme_content(github_project_url)
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=readme,
)

print(response.text)
```
